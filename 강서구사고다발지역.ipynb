```python
import pandas as pd
import folium
import re
```

```python
df = pd.read_csv('/content/drive/MyDrive/공모전/17_21_lg 강서구 사고다발.csv',encoding="cp949")
df
```

```python
df.columns
```

```python
df1 = df[['동','지점','세부지역','사고건수','사상자수','경도','위도']]
df1
```

```python
df = pd.DataFrame(df1)
df
```

```python
k = df.groupby('세부지역')['사고건수','사상자수'].sum()
k = pd.DataFrame(k)
k.sort_values(['사고건수','사상자수'],inplace=True, ascending=False)
k
```

```python
import folium
```

```python
center = [37.5508640 , 126.849650,]
mapdata = folium.Map(location=center, zoom_start = 11)

for temp in df.index:
    folium.Circle(
    [df['위도'][temp], df['경도'][temp]],
    radius = int(df.loc[temp, ['사고건수']].values[0] * 8), color='red', fill_color='red'
    
    ).add_to(mapdata)
mapdata
```
