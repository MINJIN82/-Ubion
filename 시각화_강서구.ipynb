```python
import pandas as pd
import folium
```

```python
df1=pd.read_excel("/content/drive/MyDrive/공모전/불법주정차_cctv공통_20_21_sum_위겨.xlsx")
df1
```

```python
df3=pd.read_excel("/content/drive/MyDrive/공모전/전동킥보드 견인 주소별_수정.xlsx")
df3
```

```python
df3 = df3.dropna()
df3
```

```python
df# 지도의 중심 위치를 계산합니다.
center_lat, center_lng = ( df2['위도'].mean() + df3['위도'].mean()) / 2, (df2['경도'].mean() + df3['경도'].mean()) / 2

# folium 지도 객체를 생성합니다.
m = folium.Map(location=[center_lat, center_lng], zoom_start=12)


# 두 번째 파일의 데이터프레임의 각 행에 대해 마커를 추가합니다.
for index, row in df2.iterrows():
    lat, lng = row['위도'], row['경도']
    folium.Marker(location=[lat, lng], icon=folium.Icon(icon="cirlce",color="green"), marker_type="dot").add_to(m)

# 세세 번째 파일의 데이터프레임의 각 행에 대해 마커를 추가합니다.
for index, row in df3.iterrows():
    lat, lng = row['위도'], row['경도']
    folium.Marker(location=[lat, lng], icon=folium.Icon(icon="cirlce",color="blue"), marker_type="dot").add_to(m)

m
```
